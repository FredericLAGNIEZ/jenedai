```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier
df = pd.read_csv("extract_cvs_engis_dataset.csv", sep=";")

# Convertir la date
df["date"] = pd.to_datetime(df["date"])

# Agrégation
agg = (
    df.groupby(["date", "region"])["total_energie_soutiree_wh"]
    .sum()
    .reset_index()
)

# Pivot pour tracer les régions
pivot = agg.pivot(index="date", columns="region", values="total_energie_soutiree_wh")

# Graphique
pivot.sort_index().plot()
plt.xlabel("Date")
plt.ylabel("Total énergie soutirée (Wh)")
plt.title("Consommation totale par région")
plt.show()
```
